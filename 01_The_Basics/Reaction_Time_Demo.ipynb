```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

# 1. Create fake psychology data (Reaction Times)
# We are simulating 100 participants
np.random.seed(42)
data = {
    'Participant_ID': range(1, 101),
    'Reaction_Time_ms': np.random.normal(loc=350, scale=50, size=100) # Mean 350ms, SD 50ms
}

df = pd.DataFrame(data)

# 2. visual settings to make it look scientific
sns.set_theme(style="whitegrid")
plt.figure(figsize=(10, 6))

# 3. The Plot
sns.histplot(df['Reaction_Time_ms'], kde=True, color='purple', bins=15)

# 4. Labels
plt.title('Distribution of Participant Reaction Times (N=100)', fontsize=14)
plt.xlabel('Reaction Time (ms)')
plt.ylabel('Frequency')
plt.axvline(x=350, color='red', linestyle='--', label='Mean RT (350ms)')
plt.legend()

plt.show()
```
